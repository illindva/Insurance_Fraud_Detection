#### **2.2.3** Identify and drop rows where features have illogical or invalid values, such as negative values for features that should only have positive values <font color="red">[1 Mark]</font>

```python
# Identify and drop rows where features have illogical or invalid values, such as negative values for features that should only have positive values

# Get all numerical columns dynamically
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

# Check for negative values in all numerical columns
initial_rows_illogical = df.shape[0]
for col in numerical_cols:
    if (df[col] < 0).any():
        negative_count = (df[col] < 0).sum()
        print(f"Found {negative_count} negative values in column '{col}'.")
```

Also, checking for '?' in categorical columns and treat them as missing or unknown.

The value_counts output for 'collision_type', 'property_damage', 'police_report_available' shows '?' as a category. Let's replace '?' with NaN and then potentially impute or drop.

For simplicity in this case, we'll treat them as a separate category or impute later if needed.

```python
# For now, let's just make them consistent.
for col in ['collision_type', 'property_damage', 'police_report_available']:
    if col in df.columns:
        df[col] = df[col].replace('?', np.nan)
        print(f"Replaced '?' with NaN in column: {col}")

# After replacing '?', let's re-examine if there are any new nulls from '?' replacement
# and decide to fill or drop based on overall strategy. Given the context, we'll keep them as NaN for now,
# as one-hot encoding will handle them as a distinct category if not filled.
# For simplicity of this notebook, we won't impute these NaNs here as they represent 'unknown' status.
```

#### **2.2.4** Identify and remove columns where a large proportion of the values are unique or near-unique, as these columns are likely to be identifiers or have very limited predictive power <font color="red">[1 Mark]</font>

```python
# Calculate cardinality ratio for each column
cardinality_ratios = {}
total_rows = df.shape[0]

for col in df.columns:
    unique_values = df[col].nunique()
    cardinality_ratio = unique_values / total_rows
    cardinality_ratios[col] = cardinality_ratio

# Find columns with very high cardinality (indicating potential identifiers)
# Set threshold e.g. 80% unique values or more
cardinality_threshold = 0.8
high_cardinality_cols = [col for col, ratio in cardinality_ratios.items()
                         if ratio > cardinality_threshold]

for col in high_cardinality_cols:
    if col in df.columns:
        #df.drop(columns=[col], inplace=True)
        print(f"'{col}' is considered with high cardinality (ratio: {cardinality_ratios[col]:.2f})")
```

```python
# After finding the Cardinality, below are the columns can be dropped
columns_to_drop_identifiers = ['policy_number', 'incident_location', 'insured_zip']
for col in columns_to_drop_identifiers:
    if col in df.columns:
        df.drop(columns=[col], inplace=True)
        print(f"Dropped column '{col}' due to high cardinality/identifier nature.")

# Check the dataset after these drops
print("\nDataset shape after dropping identified redundant columns:", df.shape)
print("Columns after initial data cleaning and dropping:")
print(df.columns.tolist())
```

### **6.1 Perform resampling** <font color="red">[3 Marks]</font>

Handle class imbalance in the training data by applying resampling technique.

Use the **RandomOverSampler** technique to balance the data and handle class imbalance. This method increases the number of samples in the minority class by randomly duplicating them, creating synthetic data points with similar characteristics. This helps prevent the model from being biased toward the majority class and improves its ability to predict the minority class more accurately.

**Note:** You can try other resampling techniques to handle class imbalance

#### **7.3.1** Plot ROC Curve  to visualise the trade-off between true positive rate and false positive rate across different classification thresholds <font color="red">[2 Marks]</font>

```python
# Import libraries or function to plot the ROC curve
from sklearn.metrics import roc_curve, roc_auc_score

# Define ROC function
def draw_roc(actual, probs):
    fpr, tpr, thresholds = roc_curve(actual, probs)
    auc_score = roc_auc_score(actual, probs)
    plt.figure(figsize=(7, 5))
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (1 - Specificity)')
    plt.ylabel('True Positive Rate (Sensitivity)')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()
    return thresholds, fpr, tpr, auc_score
```

```python
# Call the ROC function
thresholds_lr_train, fpr_lr_train, tpr_lr_train, auc_score_lr_train = draw_roc(y_train_pred_df_lr['Actual'], y_train_pred_df_lr['Predicted_Prob'])
print(f"\nAUC Score for Logistic Regression on training data: {auc_score_lr_train:.4f}")
```

```python
# Create columns with different probability cutoffs to explore the impact of cutoff on model performance
cutoffs = np.linspace(0.01, 0.99, 100) # Generate 100 cutoff values
for cutoff in cutoffs:
    y_train_pred_df_lr[f'Predicted_C_{cutoff:.2f}'] = y_train_pred_df_lr['Predicted_Prob'].apply(lambda x: 1 if x > cutoff else 0)

print("\nExample of predictions at different cutoffs:")
print(y_train_pred_df_lr[[f'Predicted_C_{c:.2f}' for c in [0.2, 0.5, 0.8]]].head())
```

#### **7.3.3** Plot accuracy, sensitivity, specificity at different values of probability cutoffs <font color="red">[2 Marks]</font>

```python
# Create a DataFrame to see the values of accuracy, sensitivity, and specificity at different values of probability cutoffs
cutoff_df_lr = pd.DataFrame(columns=['cutoff', 'accuracy', 'sensitivity', 'specificity'])

for cutoff in cutoffs:
    predicted_labels = y_train_pred_df_lr['Predicted_Prob'].apply(lambda x: 1 if x > cutoff else 0)
    cm = metrics.confusion_matrix(y_train_pred_df_lr['Actual'], predicted_labels)
    TN, FP, FN, TP = cm.ravel()

    accuracy = (TP + TN) / (TP + TN + FP + FN)
    sensitivity = TP / (TP + FN) if (TP + FN) > 0 else 0
    specificity = TN / (TN + FP) if (TN + FP) > 0 else 0

    cutoff_df_lr = pd.concat([cutoff_df_lr, pd.DataFrame({'cutoff': [cutoff], 'accuracy': [accuracy], 'sensitivity': [sensitivity], 'specificity': [specificity]})], ignore_index=True)

print("\nFirst 5 rows of Cutoff Analysis DataFrame:")
print(cutoff_df_lr.head())
```

```python
# Find the optimal cutoff where sensitivity and specificity are balanced (or based on business objective)
# Visually inspect the plot, or find the point where sensitivity and specificity lines cross (approximately)
# For fraud detection, we often prioritize sensitivity (recall) to minimize false negatives (missed fraud).
# Let's find the cutoff where sensitivity and specificity are closest.
optimal_cutoff_lr = cutoff_df_lr.loc[(cutoff_df_lr['sensitivity'] - cutoff_df_lr['specificity']).abs().argsort()[:1]]['cutoff'].values[0]
print(f"\nOptimal cutoff for Logistic Regression (where sensitivity and specificity are closest): {optimal_cutoff_lr:.4f}")
```
